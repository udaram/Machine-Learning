```python
import pandas as pd
```

```python
filename ='melb_data.csv'

data = pd.read_csv(filename)

data.describe()
```

```python
data.columns
```

```python
#dropna drops missing values(non available values)
data = data.dropna(axis = 0)
```

```python
y = data.Price
```

```python
#selection of features for prediction of price
features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
```

```python
x = data[features]
```

```python
print(x.head())
```

```python
x.describe()
```

```python
#code 
from sklearn.tree import DecisionTreeRegressor

data_model = DecisionTreeRegressor(random_state=1) # choose any random state
#fit the model
data_model.fit(x,y)
```

```python
print("prediction of first five houses" )
print(x.head())
print("predictiions are")
#new = [2,1,156,-37.7,144.884]
#new = new[:]
#new = np.atleast_2d(new)
print(data_model.predict(x.head()))
```

```python
from sklearn.metrics import mean_absolute_error,accuracy_score
predicted_price = data_model.predict(x)
mean_absolute_error(y,predicted_price)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y, random_state = 0)

data_model = DecisionTreeRegressor(random_state=0)

data_model.fit(x_train,y_train)
y_predict = data_model.predict(x_test)
mean_absolute_error(y_test,y_predict)
```
